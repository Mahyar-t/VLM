# NiluLab 🪷 - Google Colab Deployment

This notebook allows you to run NiluLab for free using a T4 GPU and ngrok.

### 1. Requirements
- **Runtime Type**: T4 GPU (Runtime > Change runtime type > T4 GPU)
- **ngrok Token**: You can either paste it below OR save it in the Colab **Secrets** (key icon 🔑) as `NGROK_TOKEN` for one-click startup.

```python
# @title 🚀 Start NiluLab
USE_GOOGLE_DRIVE = True # @param {type:"boolean"}
NGROK_TOKEN = "" # @param {type:"string"}
NGROK_PORT = 8080 # @param {type:"integer"}

import os
from google.colab import drive, userdata

# 1. Handle Secrets
if not NGROK_TOKEN:
    try:
        NGROK_TOKEN = userdata.get('NGROK_TOKEN')
    except:
        pass

# 2. Handle Google Drive
PROJECT_ROOT = "/content/VLM"
if USE_GOOGLE_DRIVE:
    drive.mount('/content/drive')
    PROJECT_ROOT = "/content/drive/MyDrive/NiluLab/VLM"
    os.makedirs(os.path.dirname(PROJECT_ROOT), exist_ok=True)

# 3. Clone or Move
if not os.path.exists(PROJECT_ROOT):
    print("Cloning repository to persistent storage...")
    if USE_GOOGLE_DRIVE:
        os.makedirs(PROJECT_ROOT, exist_ok=True)
        !git clone https://github.com/Mahyar-t/VLM.git "{PROJECT_ROOT}"
    else:
        !git clone https://github.com/Mahyar-t/VLM.git

%cd "{PROJECT_ROOT}"

# 4. Run Setup
!python3 scripts/colab_setup.py --token {NGROK_TOKEN} --port {NGROK_PORT} --persistent
```
